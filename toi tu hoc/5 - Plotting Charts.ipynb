```python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres, color="red", linestyle='dashed')

plt.show()
```

```python
 y_highres
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres, color="red", linestyle='dashed',  linewidth=8)

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x_highres, y_highres,color="red", linestyle='dashed',  linewidth=3, marker='o',
         markerfacecolor='blue', markersize=5)
```

```python
axes.set_title('$y=x^2$')
               
```

```python
fig.add_subplot(111).grid() #ke nhung o vuong nho tren bieu do
```

```python
axes.set_xlabel('x')
axes.set_ylabel('y')

plt.show()
```
