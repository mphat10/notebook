```python
# 1.1 Import tensorflow and other libraries.
import tensorflow as tf
import numpy as np

%matplotlib inline
import pylab
```

# DATASET

```python
# 1.2 Create input data using NumPy. y = x * 0.1 + 0.3 + noise (create dataset)
x_data = np.random.rand(100).astype(np.float32)
noise = np.random.normal(scale=0.01, size=len(x_data))
y_data = x_data * 0.1 + 0.3 + noise

# Uncomment the following line to plot our input data.
pylab.plot(x_data, y_data, '.')
```

# Bi AN

```python
# 1.3 Buld inference graph.
# Create Variables W and b that compute y_data = W * x_data + b
W = tf.Variable(tf.random_uniform([1], 0.0, 1.0)) #(shape, minval=0, maxval)  
b = tf.Variable(tf.zeros([1])) #tao bien tensor ([1])
y = W * x_data + b

# Uncomment the following lines to see W and b are.
print(W)
print(b)
```

```python
u= tf.random_uniform([1], 0.0, 1.0)# The generated values follow a uniform distribution(Phân phối đều) in the range
#`[minval, maxval)`. The lower bound `minval` is included in the range, while
#the upper bound `maxval` is excluded.
print(u)
```

```python
# 1.4 Build training graph.
loss = tf.reduce_mean(tf.square(y - y_data))  # tf.reduce_mean(input_tensor) Create an operation that calculates loss.
#^^Computes the mean (trung binh) of elements across dimensions of a tensor
optimizer = tf.train.GradientDescentOptimizer(0.5)  # Create an optimizer.
train = optimizer.minimize(loss)  # Create an operation that minimizes loss.
tf.initialize_all_variables = tf.initialize_all_variables()  # Create an operation initializes all the variables.

# Uncomment the following 3 lines to see what 'loss', 'optimizer' and 'train' are.
print("loss:", loss)
print("optimizer:", optimizer)
print("train:", train)
print(tf.initialize_all_variables)
```

```python
tf.square(y - y_data) #Di= y - y_data (tich phan Computes square 
```

```python
# 1.5 Uncomment the following line to see what we have built.
#print(tf.get_default_graph().as_graph_def())
```

```python
# 1.6 Create a session and launch the graph.
sess = tf.Session()
sess.run(tf.initialize_all_variables)
y_initial_values = sess.run(y)  # Save initial values for plotting later.

# Uncomment the following line to see the initial W and b values.
print(sess.run([W, b]))
```

```python
# tien hanh thuc tap
```

```python
# 1.7 Perform training.
for step in range(2000):
    sess.run(train)
    # Uncomment the following two lines to watch training happen real time.
    if step % 150 == 0:
        print(step, sess.run([W, b]))

print(sess.run([W, b]))
```

```python
# 1.8 Uncomment the following lines to compare.
pylab.plot(x_data, y_data, '.', label="real_values") #target value
pylab.plot(x_data, y_initial_values, ".", label="initial_values") #gia tri thay the
pylab.plot(x_data, sess.run(y), ".", label="trained_values")
pylab.legend()
pylab.ylim(0, 0.8)
```
